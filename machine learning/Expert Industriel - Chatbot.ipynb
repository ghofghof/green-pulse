## üî∂ Step 1: Data Preparation and Merging

```python
import pandas as pd
import os

# R√©pertoire contenant vos fichiers CSV
repertoire_csv = 'projet'

# Lister les fichiers CSV dans le r√©pertoire
fichiers_csv = [f for f in os.listdir(repertoire_csv) if f.endswith('.csv')]

# Dictionnaire pour stocker les DataFrames
dataframes = {}

# Charger chaque fichier CSV dans un DataFrame
for fichier in fichiers_csv:
    # Construire le chemin complet du fichier
    chemin_fichier = os.path.join(repertoire_csv, fichier)
    try:
        # Charger le CSV dans un DataFrame
        df = pd.read_csv(chemin_fichier, sep=';', encoding='utf-8', skipinitialspace=True)
        # Ajouter le DataFrame au dictionnaire
        nom_table = os.path.splitext(fichier)[0]
        dataframes[nom_table] = df
        # Afficher les premi√®res lignes du DataFrame
        print(f"Table '{nom_table}' charg√©e avec succ√®s:")
        display(df.head())  # Utilisez display() pour une meilleure pr√©sentation dans Jupyter
        print("\n")
    except Exception as e:
        print(f"Erreur lors du chargement de '{fichier}': {e}")
```

```python
# Raccourcis pour les tables importantes
fact_co2 = dataframes['Fact_Co2_Emissions']
fact_mm_pm = dataframes['Fact_MM_PM']
equipements = dataframes['Dim_Equipment']
fournisseurs = dataframes['Dim_Supplier']
materiaux = dataframes['Dim_Material']
energie = dataframes['Dim_Type_Energy']
transport = dataframes['Dim_Transport_Type']
type_mm = dataframes['Dim_Type_MM']
cat_eq=dataframes['Dim_Category_Equipment']

# Fusion avec les dimensions pertinentes pour enrichir les donn√©es CO2
df_merged = fact_co2.merge(equipements, left_on='Fk_Equipment', right_on='Pk_Equipment', how='left')
df_merged = df_merged.merge(energie, left_on='Fk_Energy', right_on='Pk_energy', how='left')
df_merged = df_merged.merge(fournisseurs, left_on='Fk_Supplier', right_on='Pk_Supplier', how='left')
df_merged = df_merged.merge(transport, left_on='Fk_Type_Transport', right_on='Pk_transport', how='left')
df_merged = df_merged.merge(materiaux, left_on='Fk_Material', right_on='Pk_Material', how='left')
df_merged = df_merged.merge(type_mm, left_on='Fk_Type_MM', right_on='Pk_Type_MM', how='left')
df_merged = df_merged.merge(cat_eq, left_on='Fk_Category', right_on='pk_category', how='left')

# Ajout de la table Fact_MM_PM : m√™me √©quipement et m√™me mati√®re
df_merged = df_merged.merge(
    fact_mm_pm,
    on=['Fk_Equipment', 'Fk_Material'],
    how='left',
    suffixes=('', '_mm_pm')  # pour √©viter les collisions de colonnes
)
print("‚úÖ Donn√©es enrichies AVEC consommation mati√®re/PM pr√™tes pour l'analyse.")
display(df_merged.head())

print("‚úÖ Donn√©es enrichies pr√™tes pour l'analyse.")
display(df_merged.head())
```

```python
# Raccourcis pour les tables suppl√©mentaires
maintenance_cycle = dataframes['Dim_Maintenance_Cycle']
category_equipment = dataframes['Dim_Category_Equipment']

# Fusionner les donn√©es
df_merged = df_merged.merge(
    maintenance_cycle,
    left_on='Fk_MCycle',  # Cl√© √©trang√®re dans Dim_Equipment
    right_on='Pk_MCycle',  # Cl√© primaire dans Dim_Maintenance_Cycle
    how='left'
)

df_merged = df_merged.merge(
    category_equipment,
    left_on='Fk_Category',  # Cl√© √©trang√®re dans Dim_Equipment
    right_on='pk_category', # Cl√© primaire dans Dim_Category_Equipment
    how='left'
)

# V√©rifier les colonnes disponibles
print("Colonnes dans df_merged :", df_merged.columns.tolist())
```

```python
df_merged = df_merged.rename(columns={
    "Category_y": "Category_Equipment"
})
colonnes_a_corriger = [
    "Renewable_Energy_Percentage",
    "CO2_Emissions_kg_equipment",
    "Carbon_Foot_Print_per_Unit_kgCo2",
    "Water_Consumption_PerUnit_L",
    "Energy_Consumption_kWh",
    "Temperature", 
    "Wind_Speed", 
    "CO2_Ambient_Level", 
    "Energy_Consumption_kWh",
    "Transport_Distance_Km"
]

for col in colonnes_a_corriger:
    df_merged[col] = df_merged[col].astype(str).str.replace(",", ".").astype(float)
```

```python
# Calculer Transport_CO2 (adapt√© de votre DAX)
def calculate_transport_co2(row):
    transport_factors = {
        1: 0.1,    # Unknown
        2: 0.03,   # Ferroviaire
        3: 0.1,    # Multimodal
        4: 0.15,   # A√©rien
        5: 0.05,   # Maritime
        6: 0.2     # Camionnage
    }
    transport_type = row['Fk_Type_Transport']
    distance = row['Transport_Distance_Km']
    return distance * transport_factors.get(transport_type, 0)

df_merged['Transport_CO2'] = df_merged.apply(calculate_transport_co2, axis=1)
```

```python
# Eau consomm√©e par unit√© produite
df_merged['Water_Per_Unit'] = df_merged.apply(
    lambda row: row['Water_Consumption_PerUnit_L'] / row['Total_Stock']
    if row['Total_Stock'] > 0 else 0, axis=1)

# CO2 par kWh
df_merged['CO2_kg_per_kWh'] = df_merged.apply(
    lambda row: row['CO2_Emissions_kg_equipment'] / row['Energy_Consumption_kWh']
    if row['Energy_Consumption_kWh'] > 0 else 0, axis=1)

# CO2 par unit√© de mati√®re (pour une marque sp√©cifique)
df_merged['CO2_Per_Unit_Material_kg'] = df_merged.apply(
    lambda row: row['Carbon_Foot_Print_per_Unit_kgCo2'] / row['Total_Stock']
    if row['Fk_Brand'] == 4 and row['Total_Stock'] > 0 else 0, axis=1)

# Score de durabilit√© fournisseur (plus le score est √©lev√©, mieux c'est)
df_merged['Supplier_Sustainability_Score'] = df_merged.apply(
    lambda row: max(0, (1 - (
        row['Transport_CO2'] / (
            row['CO2_Emissions_kg_equipment'] +
            row['Carbon_Foot_Print_per_Unit_kgCo2'] +
            row['CO2_Ambient_Level']
        )) * 100)
    ) if row['Fk_Brand'] == 4 and (
        row['CO2_Emissions_kg_equipment'] +
        row['Carbon_Foot_Print_per_Unit_kgCo2'] +
        row['CO2_Ambient_Level']
    ) > 0 else 0,
    axis=1)
```
